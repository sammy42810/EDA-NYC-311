```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Load the CSV file (adjust path as needed)
file_path = '311_Service_Requests_2024.csv'  # make sure this is in the same folder or give full path
df = pd.read_csv(file_path, dtype={8: str}, low_memory = False)

# Step 3: Preview the data
df.head()
```

```python
#Basic Data Cleansing & Wrangling

# Check for missing values
print("Missing values per column:\n", df.isnull().sum())

# Drop rows with all missing values
df = df.dropna(how='all')
print(df.columns)
df = df.drop('Agency Name', axis=1, errors='ignore')  # Drop 'Agency Name' if it exists because on the NYC 311 website, it says that this column is incorrect and should not be used

# Remove duplicate rows
df = df.drop_duplicates()

# Formats all date instances to the same datetime format
for col in df.columns:
    if 'date' in col.lower():
        df[col] = pd.to_datetime(df[col], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Save cleaned DataFrame as Parquet
df.to_parquet('311_Service_Requests_2024_cleaned.parquet', index=False)
```
